```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Salary_dataset.csv")
```

```python
df.head()
```

```python
sns.regplot(data=df, x = 'YearsExperience', y = 'Salary')
```

```python
X = df['YearsExperience']
y = df['Salary']
```

```python
np.polyfit(X,y,deg=1)
```

```python
salaryy = np.linspace(0,500,100)
```

```python
predicted_salary = 9449.96232146*salaryy + 24848.20396652
```

```python
sns.scatterplot(data=df, x='YearsExperience', y='Salary')
plt.plot(salaryy, predicted_salary,color='pink')
```
